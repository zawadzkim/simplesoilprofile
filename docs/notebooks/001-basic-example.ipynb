```python
from simplesoilprofile import SoilLayer, SoilProfile, plot_profile

import matplotlib.pyplot as plt
from shapely.geometry import Point

# Create soil layers
topsoil = SoilLayer(
    name="Topsoil",
    theta_res=0.02,
    theta_sat=0.4,
    alpha=0.02,
    n=1.5,
    k_sat=10.0,
    texture_class="sandy loam",
    clay_content=10.0,
    silt_content=20.0,
    sand_content=70.0,
)

subsoil = SoilLayer(
    name="Subsoil",
    theta_res=0.05,
    theta_sat=0.45,
    alpha=0.01,
    n=1.3,
    k_sat=5.0,
    texture_class="clay loam",
    clay_content=30.0,
    silt_content=35.0,
    sand_content=35.0,
)

bedrock = SoilLayer(
    name="Bedrock",
    theta_res=0.1,
    theta_sat=0.3,
    alpha=0.005,
    n=1.2,
    k_sat=1.0,
    texture_class="rock",
    clay_content=5.0,
    silt_content=10.0,
    sand_content=85.0,
)

# Create soil profile
profile = SoilProfile(
    name="Test Profile",
    layers=[topsoil, subsoil, bedrock],
    layer_depths={
        0: (0, 30),    # Topsoil from 0-30 cm
        1: (30, 100),  # Subsoil from 30-100 cm
        2: (100, 200), # Bedrock from 100-200 cm
    },
    location=Point(100.0, 200.0),
    elevation=5.0,
)

# Create plot
fig, ax = plt.subplots(figsize=(8, 12))
plot_profile(profile, ax=ax, show_properties=True)
plt.show()
```
